## Overfitting

When a model learns too much from the training data, including irrelevant information such as noise or outliers, this is known as overfitting.

Consider fitting a complex curve to a group of points, for instance.  Although the curve will pass through each point, it won't accurately depict the pattern.
Because of this, the model performs admirably on training data but poorly on fresh data.

Example: Overfitting models are comparable to classroom students who learn answers by heart or cramming rather than comprehending/ understanding the subject core concepts. They perform well on training tests, but they have trouble on actual exams.

# Underfitting

The reverse of overfitting is underfitting.  It occurs when a model is too basic to adequately represent the data.

Consider, for instance, fitting points that truly follow a curve with a straight line.  The majority of the pattern is missed by the line.
In this instance, the model performs poorly on both the testing and training data.\
Example: Underfitting models are comparable to undereducated students.  They perform poorly on both practice examinations and actual exams.

# Bias

A model is said to be biased if it frequently makes mistakes because of over basic assumptions.  Consider it the innate prejudice of a model or a basic weakness in its logic.  High bias models are unable to adequately represent the fluctuation of the data, which inevitably results in underfitting.
