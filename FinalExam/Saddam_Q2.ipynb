## 2.(a) What is Overfitting, and How Can You Avoid It?


Overfitting occurs when a machine learning model performs badly on fresh, unseen data because it has learned the training data too well, including its noise and random fluctuations.  Consider a pupil who masters all of the answers on a practice test but fails the real one due to slightly different questions.  Techniques like regularization, which penalizes complex models to promote simpler, more generalized ones, and cross-validation, which evaluates the model's performance on several subsets of the data, can help you prevent this.  Using a bigger and more varied dataset to train your model is another efficient technique.

## 2.(b) What do you understand by the term Bias-Variance trade-off?

A key idea in machine learning is the bias-variance trade-off, which examines the connection between a model's complexity and generalizability.  The term "bias" describes a model's ability to continuously make simplistic assumptions that lead to incorrect learning.  Frequently, a high-bias model underfits the data and is overly simplistic.  Conversely, variance describes how sensitive the model is to even slight changes in the training set.  Frequently, a high-variance model overfits the data and is overly complicated.  Finding the sweet spot, a model that balances bias and variance, is complex enough to capture the underlying patterns without being unduly sensitive to the noise in the training data is the aim.

## 2.(c) When Will You Use Classification over Regression?

Classification is used when you want to forecast a discrete, categorical value, like "yes" or "no," "spam" or "not spam," or "dog," "cat," or "bird."  Assigning an input to one of these preset categories is the model's responsibility.  However, regression is used when the result is a continuous, numerical number, for as when estimating an individual's age or a home's price.   The primary distinction is in the output type: regression predicts a quantity, whereas classification predicts a label.

## 2.(d) What is Cross-Validation?

A statistical method for assessing how well a machine learning model would generalize to a different dataset is called cross-validation.  By doing this, you can make sure the model isn't only "lucky" on one set of tests.  K-fold cross-validation is a popular technique that divides the dataset into k folds of equal size.  k-1 of these folds are used to train the model, and the remaining fold is used for validation.  Every fold is used as the validation set exactly once during the k repetitions of this operation.  A more reliable assessment of the model's performance is then obtained by averaging the outcomes.
